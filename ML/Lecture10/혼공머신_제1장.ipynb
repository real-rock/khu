# 인공지능이란?

## 인공지능의 분류

- **인공일반지능(Artificial General Intelligence), 강인공지능(Strong AI)**: [Her](https://movie.daum.net/moviedb/main?movieId=80976) 사만다, [Terminator](https://movie.daum.net/moviedb/main?movieId=1911) 스카이넷 
- **약인공지능 (Weak AI)**: 특정분야에서 사람의 일을 도와주는 보조 역할 
  - 예. 음성 비서, 자율 주행 자동차, 음악 추천, 기계 번역, 알파고 등

## 데이터 가공

````python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

print(length)
len(length)
````

````python
fish_data = [[l,w] for l, w in zip(length,weight)]
````

````python
fish_data[10]
````

````python
fish_data
````

````python
[1]*10+[0]*2
````

````python
fish_target = [1]*len(bream_length)+[0]*len(smelt_length)
print(fish_target)
````

## k-최근접 이웃 알고리즘 적용 



```
# scikit-learn 모듈 사용

# 객체 생성

```





````python
# scikit-learn 모듈 사용
from sklearn.neighbors import KNeighborsClassifier

# 객체 생성
kn = KNeighborsClassifier()

# 훈련 
kn.fit(fish_data, fish_target)
````

## 개발 모델 평가

* score 메소드 
* 정확한 답을 준 확률로 **정확도** 평가

````python
kn.score(fish_data,fish_target)
````

## k-최근접 이웃 알고리즘



*   어떤 데이터에 대한 답을 구할 때 주위의 다른 데이터(k개)를 보고 다수를 차지하는 것을 정답으로 사용함
*   새로운 데이터와 기존 데이터만 있으면 매우 간단하게 구현가능
*   https://www.youtube.com/watch?v=HVXime0nQeI 참고
*   https://www.youtube.com/watch?v=k_7gMp5wh5A 참고




````python
kn.predict([[10,20]])
````

````python
kn.n_neighbors
````

````python
new_fish = [[10,20]]

for n in range(5,50):
  kn.n_neighbors = n
  print(n,kn.predict(new_fish))
````
